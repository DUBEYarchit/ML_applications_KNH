```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_boston
```

```python
df = load_boston()
```

```python
df 
```

Creating a pandas DataFrame from the data portion of a scikit-learn dataset

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head(10)
```

Independent and Dependent Features

```python
X = dataset

y = df.target
```

```python
# Price of house 
y
```

```python
# train test split 
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.3 , random_state=42)
```

```python
X_train 
```

```python
# Standatdizing Dataset 

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

# Cross - validation 

from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()

regression.fit(X_train,y_train)
```

```python
mse = cross_val_score (regression, X_train, y_train, scoring = 'neg_mean_squared_error', cv = 5)
```

```python
np.mean(mse)
```

## Prediction 

```python
pred = regression.predict(X_test)
pred
```

To see the difference between the predicted values by the model and the actual y_test values 

making a density plot 

```python
import seaborn as sns 

sns.kdeplot(pred - y_test)
```

Less variance from -10 to + 10 thus model is good
