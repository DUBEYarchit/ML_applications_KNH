Selecting a small dataset for post prunning as it is more effective with small dataset.

```python
iris = load_iris()
```

```python
import seaborn as sns
```

```python
sns.load_dataset('iris')
```

```python
df = sns.load_dataset('iris')
```

## Independent and dependent features 

```python
X = df.iloc[:,:-1]
X
```

```python
y = iris.target
y
```

### Train Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# post prunning as 1st we are making tree with all the pre determined parameters of DecisionTreeClassifier()
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train,y_train)
```

```python
# Making a plot to see the Decision Tree 

from sklearn import tree

plt.figure(figsize=(14,14))
tree.plot_tree(tree_model,filled=True)
```

Now it is evident that after the depth of 2 there are 2 majority classes in the nodes which are 'versicolor' and 'virginica'

To avoid overfitting here we can set the parameter of max_depth as 2 

```python
tree_model = DecisionTreeClassifier(max_depth=2)
```

```python
tree_model.fit(X_train,y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(14,14))
tree.plot_tree(tree_model,filled=True)
```

Making prediction with the model after post prunning 

```python
y_predict = tree_model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_predict,y_test))
```

```python
print(classification_report(y_predict,y_test))
```
