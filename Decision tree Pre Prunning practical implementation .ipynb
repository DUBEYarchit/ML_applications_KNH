```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
data.data
```

```python
import seaborn as sns
```

```python
pre_df = sns.load_dataset('iris')
pre_df
```

### Independent and Target variable

```python
X = pre_df.iloc[:,:-1]
X
```

```python
y = data.target
y
```

### Train Test Split with sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

### Pre - Prunning 

```python
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_pre = DecisionTreeClassifier()
new_model= GridSearchCV(tree_pre,param_grid=parameter,cv=6,scoring='accuracy')
```

```python
new_model.fit(X_train,y_train)
```

```python
new_model.best_params_
```

```python
y_test
```

```python
y_pred_pre = new_model.predict(X_test)
y_pred_pre
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_pred_pre,y_test)
score
```

```python
print(classification_report(y_pred_pre,y_test))
```
